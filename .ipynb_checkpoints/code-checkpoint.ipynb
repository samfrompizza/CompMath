```python
import numpy as np
from numba import prange, jit
import typing
from collections.abc import Callable
from sklearn.datasets import make_spd_matrix
from matplotlib import pyplot as plt
import seaborn as sns

sns.set(style="darkgrid", font_scale=1.4)
```

```python
f = lambda x: x**5# x*np.sin(2*x) + x
f_der = lambda x: 5*x**4# np.sin(2*x) + 2*x*np.cos(2*x) + 1
f_der_2 = lambda x: 20*x**3# 4*(np.cos(2*x)-x*np.sin(2*x))

def DifferenceScheme(Y: np.ndarray) -> np.ndarray:
    n = Y.shape[0]
    der = np.zeros(n)
    der[:-1] = (Y[1:] - Y[:-1])*(n-1)
    der[n-1] = der[n-2]
    return der
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
X = np.linspace(0, 1, 200)
ax[0].plot(X, DifferenceScheme(f(X)), label='Difference scheme')
ax[0].plot(X, f_der(X), label='Real derivative')

Npoints = np.arange(10, 10000, 100)
error_data = np.ones(Npoints.shape[0])

for i, n in enumerate(Npoints):
    X = np.linspace(0, 1, n)
    error_data[i] = np.linalg.norm(DifferenceScheme(f(X)) - f_der(X), np.inf)

ax[1].plot(Npoints, error_data, label='Error')
ax[1].plot(Npoints, 1/Npoints, label=r'$\frac{1}{N}$')
ax[1].set_yscale('log')
ax[1].set_xscale('log')

ax[0].legend()
ax[1].legend()
plt.tight_layout()
```

```python
def DifferenceScheme2(Y: np.ndarray) -> np.ndarray:
    n = Y.shape[0]
    der = np.zeros(n)
    der[1:-1] = (Y[2:] - 2*Y[1:-1] + Y[:-2])*(n-1)**2
    der[0] = (-Y[3]+4*Y[2]-5*Y[1]+2*Y[0])*(n-1)**2
    der[n-1] = (-Y[n-4]+4*Y[n-3]-5*Y[n-2]+2*Y[n-1])*(n-1)**2
    return der
```

```python
fig, ax = plt.subplots(1, 2, figsize=(20, 10))
X = np.linspace(0, 1, 2000)
ax[0].plot(X, DifferenceScheme2(f(X)), label='Difference scheme')
ax[0].plot(X, f_der_2(X), label='Real derivative')

Npoints = np.arange(10, 10000, 100)
error_data = np.ones(Npoints.shape[0])

for i, n in enumerate(Npoints):
    X = np.linspace(0, 1, n)
    error_data[i] = np.linalg.norm(DifferenceScheme2(f(X)) - f_der_2(X), np.inf)

ax[1].plot(Npoints, error_data, label='Error')
ax[1].plot(Npoints, 1/(Npoints**2), label=r'$\frac{1}{N^2}$')
ax[1].set_yscale('log')
ax[1].set_xscale('log')

ax[0].legend()
ax[1].legend()
plt.tight_layout()
```
