```python
import numpy as np
import matplotlib.pyplot as plt
```

## Формулировка задачи

Посчитать интаграл:

$\int_{0}^{1} \sin{x^2} dx $

- Методом правых квадратов
- Методом средних квадратов
- Методом Симпсона

```python
def f(x):
    return np.sin(x ** 2)  # функция для интегрирования
    
a = 0  # начало интервала
b = 1  # конец интервала
n = 20  # количество отрезков для примера
```

## Метод правых и центральных прямоугольников

```python
def right_rectangle_method(a, b, n):
    h = (b - a) / n 
    x = np.linspace(a, b - h, n)
    return np.sum(f(x + h) * h)  

def central_rectangle_method(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b - h, n)
    return np.sum(f(x + h / 2) * h) 
```

## Метод Симпсона

```python
def simpson_method(a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b - h, n)
    return h/6 * np.sum(f(x) + 4 * f(x + h / 2) + f(x + h)) 
```

```python
# Вычисление интегралов
integral_right = right_rectangle_method(a, b, n)
integral_central = central_rectangle_method(a, b, n)
integral_simpson = simpson_method(a, b, n)

print("Интеграл методом правых прямоугольников:", integral_right)
print("Интеграл методом центральных прямоугольников:", integral_central)
print("Интеграл методом Симпсона:", integral_simpson)
```

```python
# Визуализация
x = np.linspace(a, b, 1000)
y = f(x)

# Узлы для построения прямоугольников
x_rect = np.linspace(a, b - (b - a) / n, n)
h = (b - a) / n

# Построение графиков
plt.figure(figsize=(9, 6))
plt.plot(x, y, label="f(x) = sin(x)", color="blue")

# Метод правых прямоугольников
for i in range(n):
    plt.bar(x_rect[i] + h, f(x_rect[i] + h), width=h, align='edge', color='orange', edgecolor='black', alpha=0.5)

# Метод центральных прямоугольников
for i in range(n):
    plt.bar(x_rect[i] + h / 2, f(x_rect[i] + h / 2), width=h, align='center', color='green', edgecolor='black', alpha=0.5)

# Метод Симпсона
for i in range(n):
    plt.bar(x_rect[i] + h / 2, f(x_rect[i] + h / 2), width=h, align='center', color='green', edgecolor='black', alpha=0.5)

plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x) = sin(x)", "Правые прямоугольники", "Центральные прямоугольники"])
plt.title("Методы численного интегрирования: Правые и Центральные прямоугольники")
plt.show()
```
